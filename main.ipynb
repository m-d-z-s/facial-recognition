```python
from fer import FER
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
angry_image = plt.imread('photos/angry.jpeg')
angry_image2 = plt.imread('photos/angry2.jpeg')
happy_image = plt.imread('photos/happy.jpeg')
happy_image2 = plt.imread('photos/happy2.jpeg')
neutral_image = plt.imread('photos/neutral.jpeg')
neutral_image2 = plt.imread('photos/neutral2.jpeg')
sad_image = plt.imread('photos/sad.jpeg')
sad_image2 = plt.imread('photos/sad2.jpeg')
surprise_image = plt.imread('photos/surprise.jpeg')
surprise_image2 = plt.imread('photos/surprise2.jpeg')
images = [angry_image, angry_image2, happy_image, happy_image2, neutral_image, neutral_image2, sad_image, sad_image2, surprise_image, surprise_image2]
```

```python
emo_detector = FER(mtcnn=True)
```

```python
detected_emotions = []
for image in images:
    detected_emotions.append((emo_detector.detect_emotions(image), emo_detector.top_emotion(image)))
```
